```python
import numpy as np
import cPickle as pickle
from math import sqrt
from pybrain.datasets.supervised import SupervisedDataSet as SDS
from pybrain.tools.shortcuts import buildNetwork
from pybrain.supervised.trainers import BackpropTrainer
```

```python
train_path = "../cleanNeuralNetsData/train.csv"
train_target_path = "../cleanNeuralNetsData/train_target.csv"
test_path = "../cleanNeuralNetsData/test.csv"
test_target_path = "../cleanNeuralNetsData/test_target.csv"
output_model_file_path = 'model.pkl'
```

```python
hidden_size = 100
epochs = 10
```

```python
# load data
train = np.loadtxt( train_path, delimiter = ',', skiprows=1 )
train_target = np.loadtxt( train_target_path, delimiter= ",", skiprows=1)
test = np.loadtxt( test_path, delimiter = ',', skiprows=1)
test_target = np.loadtxt( test_target_path, delimiter= ",", skiprows=1 )
```

```python
train_target.shape
```

```python
train_target.shape
```

```python
train_target = train_target.reshape(-1, 1)
test_target = test_target.reshape(-1, 1)
```

```python
train_target.shape
```

```python
train.shape[1]
```

```python
input_size = train.shape[1]
target_size = train_target.shape[1]
```

```python
input_size
```

```python
target_size
```

```python
train.shape
```

```python
train_target.shape
```

```python
# prepare dataset
ds = SDS( input_size, target_size )
ds.setField( 'input', train )
ds.setField( 'target', train_target )
```

```python
# init and train
net = buildNetwork( input_size, hidden_size, target_size, bias = True )
trainer = BackpropTrainer( net,ds )
```

```python
print "training for {} epochs...".format( epochs )
```

```python
for i in range( epochs ):
	mse = trainer.train()
	rmse = sqrt( mse )
	print "training RMSE, epoch {}: {}".format( i + 1, rmse )
```

```python
pickle.dump( net, open( output_model_file, 'wb' ))
```

```python
trainer.testOnData(verbose=True)
```
